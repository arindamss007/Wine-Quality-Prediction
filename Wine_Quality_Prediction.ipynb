```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

wine_df = pd.read_csv('/content/winequalityN.csv')

wine_df
```

**COLUMNS INFORMATION**

```python
wine_df.info()
```

**DATA CLEANING PHASE**

```python
col = ['fixed acidity',
       'volatile acidity',
       'citric acid',
       'residual sugar',
       'chlorides',
       'free sulfur dioxide',
       'total sulfur dioxide',
       'density',
       'pH',
       'sulphates',
       'alcohol',
       'quality', ]

i=0
for column in col :
  meanval = wine_df[column].mean()
  print(meanval)
  wine_df.fillna(meanval, inplace=True)
```

**COLUMN DESCRIPTION**

```python
wine_df.describe()
```

**CHANGING DATA TYPE**

```python
wine_df['quality'] = wine_df['quality'].astype(int)
wine_df.info()
```

**VISUALIZATION FOR WINE TYPES**

```python
sns.countplot(x=wine_df['type'])
```

**CORRELATION OF COLUMNS**

```python
sns.heatmap(wine_df.corr(), annot=True, linewidths=5)
plt.show()
```

**FEATURE EXTRACTION FOR ML MODEL**

```python
features = ['fixed acidity',
       'volatile acidity',
       'citric acid',
       'residual sugar',
       'chlorides',
       'free sulfur dioxide',
       'total sulfur dioxide',
       'density',
       'pH',
       'sulphates',
       'alcohol', ]
target = wine_df['quality']
```

**COLUMNS RELATION WITH QUALITY**

```python
i = 0

for column in features :
  plt.subplot(4,3, i+1)
  i=i+1
  plt.bar(target, wine_df[column])
  plt.xlabel('Quality')
  plt.ylabel(column)
  plt.tight_layout()
```

**FEATURE ENGINEERING FOR DATA**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit_transform(wine_df[features])
```

**BUILDING REGRESSION MODELS FOR PREDICTION**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression, SGDRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
import warnings

x_train, x_valid, y_train, y_valid = train_test_split(wine_df[features], wine_df['quality'], train_size=0.80)

models = [LinearRegression(),
          LogisticRegression(max_iter=1000, tol=0.001, random_state=42,),
          DecisionTreeRegressor(max_depth=15, max_features=11, random_state=42),
          RandomForestRegressor(n_estimators=1000, max_depth=15),
          SVR(max_iter=1000), ]
modelname =['Multiple Regression', 'Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM Regression']
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_valid.shape)
print(y_valid.shape)
```

**GETTING PREDICTION RESULT OF OUR MODELS**

```python
i=0
for model in models :
  print(f'{modelname[i]} Algorithm Report : \n')
  i = i+1
  wine_model = model.fit(x_train, y_train)
  preds = wine_model.predict(x_valid)
  print(f'Mean Absolute Error : {mean_absolute_error(y_valid, preds)}')
  print(f'Accuracy Score : {r2_score(y_valid, preds)}')
  print("\n")
```

**HERE WE MADE A SERIES OF BINARY DATA FOR 7+ RATING WINES**

```python
import numpy as np
import pandas as pd

result = [] #EMPTY LIST FOR HOLDING BINARY VALUES

r = wine_df['quality'].shape
quality = wine_df['quality']

for i in range(0,r[0]) :
    if(quality[i]>=7):
        result.append(1)
    else :
        result.append(0)

column = ['result']
res_df = pd.DataFrame(result, columns=column)
wine_df
```

**BINARY DATA COLUMN INFORMATION**

```python
res_df.info()
```

**BINARY COLUMN WHETHER THE WINE'S RATING IS 7 OR OVER 7, COSIDERED AS GOOD WINE**

```python
res_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC, LinearSVC
import sklearn.metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

x_train, x_valid, y_train, y_valid = train_test_split(wine_df[col], res_df['result'], train_size=0.80)

models = [ DecisionTreeRegressor(max_depth=15, max_features=13, random_state=42), ]
modelname =['DecisionTree Classifier', ]
```

**GETTING CLASSIFICATION REPORT**

```python
i=0
for model in models :
  print(f'{modelname[i]} Algorithm Report : \n')
  i = i+1
  wine_model = model.fit(x_train, y_train)
  preds = wine_model.predict(x_valid)
  print(f'Mean Absolute Error : {mean_absolute_error(y_valid, preds)}')
  print(f'Accuracy Score : {r2_score(y_valid, preds)}')
  print("\n")
```

**CONFUSION MATRIX FOR FINALIZE THE MODEL**

```python
i=0

for model in models :
    print(f'{modelname[i]} Report : ')
    i=i+1
    model.fit(x_train, y_train)
    preds = model.predict(x_valid)
    cm = confusion_matrix(y_valid, preds)
    sns.heatmap(cm, annot=True)
    plt.title("Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```
